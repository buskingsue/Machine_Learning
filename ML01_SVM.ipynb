```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
iris_dataset = load_iris()
iris = pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)
labels = iris_dataset.target_names
iris.info()
iris.head()
```

```python
scaler = StandardScaler()
iris_scaled = scaler.fit_transform(iris)
iris_scaled = pd.DataFrame(iris_scaled, columns=['sl', 'sw', 'pl', 'pw'])
iris_scaled.head()
```

```python
target = iris_dataset.target
target
```

```python
x_train, x_test, y_train, y_test = train_test_split(iris_scaled, target, test_size=0.2)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn import svm
iris_svm = svm.SVC(C=1.0, class_weight='balanced', kernel='linear', gamma='scale')
iris_svm.fit(x_train, y_train)
```

```python
iris_svm.score(x_test, y_test)
```

```python
iris_svm.score(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = iris_svm.predict(x_test)
pd.DataFrame(confusion_matrix(y_test, y_pred),
             columns=['pred setosa', 'pred versicolors', 'pred virginica'],
             index=['actual setosa', 'actual versicolors', 'actual virginica'])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
